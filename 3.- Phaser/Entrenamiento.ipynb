```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load data from CSV
file_path = "Entrenamiento.csv"
df = pd.read_csv(file_path, header=None,delimiter=' ', names=['despBala', 'velocidadBala', 'despBala2', 'velocidadBala2', 'estatusSuelo', 'estatusQuieto'])
plt.figure()

# Graficar los puntos en 2D con colores diferentes para estatusSuelo=0 y estatusSuelo=1
plt.scatter(df[df['estatusSuelo'] == 0]['despBala'], df[df['estatusSuelo'] == 0]['velocidadBala'], color='blue', label='estatusSuelo=0', marker='o')
plt.scatter(df[df['estatusSuelo'] == 1]['despBala'], df[df['estatusSuelo'] == 1]['velocidadBala'], color='red', label='estatusSuelo=1', marker='x')

plt.xlabel('despBala')
plt.ylabel('velocidadBala')
plt.title("Vertical")

plt.legend()
plt.show()

plt.figure()

# Graficar los puntos en 2D con colores diferentes para estatusQuieto=0 y estatusQuieto=1
plt.scatter(df[df['estatusQuieto'] == 0]['despBala2'], df[df['estatusQuieto'] == 0]['velocidadBala2'], color='blue', label='estatusQuieto=0', marker='o')
plt.scatter(df[df['estatusQuieto'] == 1]['despBala2'], df[df['estatusQuieto'] == 1]['velocidadBala2'], color='red', label='estatusQuieto=1', marker='x')

plt.xlabel('despBala2')
plt.ylabel('velocidadBala2')
plt.title("Horizontal")

plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Cargar los datos desde el CSV
file_path = "Entrenamiento.csv"
df = pd.read_csv(file_path, header=None,delimiter=' ', names=['despBala', 'velocidadBala', 'despBala2', 'velocidadBala2', 'estatusSuelo', 'estatusQuieto'])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar puntos con estatusSuelo=0
ax.scatter(df[df['estatusSuelo'] == 0]['despBala'], df[df['estatusSuelo'] == 0]['velocidadBala'], df[df['estatusSuelo'] == 0]['estatusSuelo'],
           c='blue', marker='o', label='estatusSuelo=0')
# Graficar puntos con estatusSuelo=1
ax.scatter(df[df['estatusSuelo'] == 1]['despBala'], df[df['estatusSuelo'] == 1]['velocidadBala'], df[df['estatusSuelo'] == 1]['estatusSuelo'],
           c='red', marker='x', label='estatusSuelo=1')

# Definir etiquetas de los ejes
ax.set_xlabel('despBala')
ax.set_ylabel('velocidadBala')
ax.set_zlabel('estatusSuelo')
ax.set_title("Bala vertical")

fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')

# Graficar puntos con estatusSuelo=0
ax2.scatter(df[df['estatusQuieto'] == 0]['despBala2'], df[df['estatusQuieto'] == 0]['velocidadBala2'], df[df['estatusQuieto'] == 0]['estatusQuieto'],
           c='blue', marker='o', label='estatusQuieto=0')
# Graficar puntos con estatusSuelo=1
ax2.scatter(df[df['estatusQuieto'] == 1]['despBala2'], df[df['estatusQuieto'] == 1]['velocidadBala2'], df[df['estatusQuieto'] == 1]['estatusQuieto'],
           c='red', marker='x', label='estatusQuieto=1')

# Definir etiquetas de los ejes
ax2.set_xlabel('despBala2')
ax2.set_ylabel('velocidadBala2Y')
ax2.set_zlabel('estatusQuieto')
ax2.set_title("Bala Horizontal")

plt.show()
```
